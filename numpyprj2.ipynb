Arithmetic Operation

```python
import numpy as np
var = np.array([1,2,3,4,5])
varAdd =np.add(var,3)
print(varAdd)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])
varAdd = np.add(var1,var2)
print(varAdd)
```

```python
var1 = np.array([4,6,6,8])
var2 = np.array([1,2,3,4])
varSub = np.subtract(var1,var2)
print(varSub)
```

```python
var1 = np.array([4,6,6,8])
var2 = np.array([1,2,3,4])
varmul = np.multiply(var1,var2)
print(varmul)
```

```python
var1 = np.array([4,6,6,8])
var2= np.array([1,2,3,4])
vardiv = np.divide(var1,var2)
print(vardiv)
```

```python
var1 = np.array([2,5,6,6])
var2 = np.array([1,2,3,4])
varmod = np.mod(var1,var2)
print(varmod)
```

```python
var1 = np.array([2,5,6,6])
var2 = np.array([1,2,3,4])
varpow = np.power(var1,var2)
print(varpow)
```

```python
var1 = np.array([1,5,3,5])
var2 = np.array([1,2,3,4])
varrec = np.reciprocal(var1)
print(varrec)
```

2D array

```python
var21 = np.array([[1,2,3,4],[2,4,5,6]])
var22 = np.array([[1,2,3,4],[2,4,5,6]])
var23 = np.add(var21,var22)
print(var23)
```

Shape And Reshape of numpy Array

```python
var31 = np.array([[1,2,3,4],[2,4,5,6]])
print(var31.shape)
var32 = np.array([1,2,3,4],ndmin=4)
print(var32)
print(var32.shape)
```

```python
var41 = np.array([1,2,3,5])
x = var41.reshape(2,2)
print(x)
```

Broadcasting

```python
var51 = np.array([1,2,3])
print(var51.shape)
var52 = np.array([[1],[2],[3]])
print(var52)
print(var52.shape)
print(var51 + var52)
```
